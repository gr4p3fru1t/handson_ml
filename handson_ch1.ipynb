### 2.1 배치학습
시스템이 점진적으로 학습 불가 -> 가능한 데이터를 모두 사용해 훈련해야 함

### 2.2 온라인 학습
데이터를 순차적으로 한 개씩 또는 미니배치로 주입하여 시스템 훈련   
매 학습 단계 빠르고 비용 적게 들어 즉시 학습 가능  
연속적으로 데이터를 받고 빠른 변화에 적응하는 시스템에 적합 (예 : 주식가격)
